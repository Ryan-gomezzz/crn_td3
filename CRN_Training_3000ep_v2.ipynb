```
# Cell 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```
# Cell 2: Install dependencies and copy source files from Drive
import subprocess, shutil, os, time

from google.colab import drive
drive.mount('/content/drive', force_remount=True)
time.sleep(2)

subprocess.run(['pip', 'install', 'scipy', '-q'], check=True)

drive_dir = '/content/drive/MyDrive/CRN_Training'
for f in ['config.py','environment.py','utils.py','ddpg.py','td3.py']:
    src = f'{drive_dir}/{f}'
    dst = f'/content/{f}'
    for attempt in range(3):
        try:
            shutil.copy(src, dst)
            print(f'Copied {f}')
            break
        except OSError as e:
            print(f'Retry {attempt+1} for {f}: {e}')
            time.sleep(2)
            drive.mount('/content/drive', force_remount=True)
            time.sleep(2)
```

```
# Cell 3: Write camo_td3.py (embedded source)
import base64, os
_b64 = ''
with open('/content/camo_td3.py', 'wb') as f:
    f.write(base64.b64decode(_b64))
print(f'camo_td3.py: {os.path.getsize("/content/camo_td3.py")} bytes')
```

```
# Cell 4: Write train_compare_colab.py (embedded source)
import base64, os
_b64 = ''
with open('/content/train_compare_colab.py', 'wb') as f:
    f.write(base64.b64decode(_b64))
print(f'train_compare_colab.py: {os.path.getsize("/content/train_compare_colab.py")} bytes')
```

```
# Cell 5: Verify all files
import os
files = ['config.py','environment.py','utils.py','ddpg.py','td3.py','camo_td3.py','train_compare_colab.py']
for f in files:
    size = os.path.getsize(f'/content/{f}')
    print(f'  {f}: {size} bytes')
print('\nAll files ready!')
```

```
# Cell 6: Create output directories on Drive
import os
os.makedirs('/content/drive/MyDrive/CRN_Training/checkpoints', exist_ok=True)
os.makedirs('/content/drive/MyDrive/CRN_Training/results', exist_ok=True)
print('Directories ready')
```

```
# Cell 7: Run training - TD3 vs DDPG vs CAMO-TD3, 3000 episodes
import subprocess, sys
result = subprocess.run(
    [sys.executable, '/content/train_compare_colab.py',
     '--episodes', '3000',
     '--steps-per-ep', '200',
     '--agents', 'td3,ddpg,camo-td3',
     '--checkpoint-every', '300',
     '--checkpoint-dir', '/content/drive/MyDrive/CRN_Training/checkpoints',
     '--output', '/content/drive/MyDrive/CRN_Training/results/crn_comparison_report.pdf'],
    check=False
)
print(f'\nTraining complete. Return code: {result.returncode}')
```
